```python
import pandas as pd
```

```python
retail_1 = pd.read_excel('online_retail_II.xlsx',sheet_name='Year 2009-2010')
retail_2 = pd.read_excel('online_retail_II.xlsx',sheet_name='Year 2010-2011')

retail = pd.concat([retail_1,retail_2],ignore_index=True)
```

```python
retail.info()
```

```python
retail.head()
```

```python
# 1. Data Cleaning:

retail.isnull().sum()
```

**Handle missing values (especially CustomerID and Description)**

```python
# Since each product has a unique product code
# we can search the records and replace the missing description with the product which has the same product code

def fillDescription(col):

  code = col[0]
  description = col[1]

  if pd.isnull(description):

    for item in zip(retail['StockCode'],retail['Description']):

      if item[0] == code:

        return item[1]

  return description
```

```python
retail['Description'] = retail[['StockCode','Description']].apply(fillDescription,axis=1)
```

```python
retail.isnull().sum()
```

```python
# BEST CASE WE CAN FILL THE CUSTOMER ID WITH -1
# ESTIMATING WILL CAUSE UNWANTED RESULTS
retail['Customer ID'] = retail['Customer ID'].fillna(-1)

retail.isnull().sum()
```

```python
# droping na and duplicate rows

retail = retail.dropna().drop_duplicates()
retail.isnull().sum()
```

```python
# Handle canceled orders (where InvoiceNo starts with 'C')
# handling with 'quantity' column because many cancelled oredrs are not starting with 'C'
# Splitting data into fullfilled and cancelled orders
# so we can analyze the cancelled data if needed

a = [x<0 for x in retail['Quantity'] ]
canc_orders = retail[a]
retail = retail[[not x for x in a]]
```

```python
canc_orders.head()
```

```python
# Removing or cap outliers in Quantity and UnitPrice

import seaborn as sns

quantity = retail['Quantity']

Q1 = quantity.quantile(0.25)
Q3 = quantity.quantile(0.75)
med = quantity.median()

IQR = Q3-Q1
lower_bound = Q1 - (1.5*IQR)
upper_bound = Q3 + (1.5*IQR)

print(lower_bound,upper_bound)
```

```python
# Now that we have upper and lower bound we can cap the exceeding values
# not removing them since that will cause a lot of data loss

retail['Quantity'] = retail['Quantity'].clip(lower=lower_bound,upper=upper_bound)
# confirming that we limited the quantity

print(retail[retail['Quantity']> upper_bound].shape)
print(retail[retail['Quantity']< lower_bound].shape)
```

```python
# DATA CLEANING
# seperating from misc values in products such as postage etc

retail_clean = retail[(retail['Description'] != 'POSTAGE') &
                      (retail['Description'] != 'Manual') &
                      (retail['Description'] != 'DOTCOM POSTAGE') &
                      (retail['Description'] != 'AMAZON FEE')]

# seperating from returned orders

conc_ord = ~retail_clean['Invoice'].str.contains('C', na=False)
retail_clean = retail_clean[(retail_clean['Quantity'] > 0) & conc_ord & (retail_clean['Price'] > 0)]
```

```python
# Now that we have complete and clean data we can start
# Exploratory Data Analysis (EDA)

# How many unique products are there
# can either find from stock code or description

unique_products = retail_clean['StockCode'].unique()
print(f'No of unique products: {len(unique_products)}')
```

```python
# What is the total number of transactions

# total transaction can be calculated with uniique invoices

unique_trans = retail_clean['Invoice'].unique()

print(f"Total transaction: {len(unique_trans)}")
```

```python
# What countries do the customers belong to

countries = retail_clean['Country'].unique()
countries
```

```python
c_plot = sns.countplot(x='Country',data=retail_clean)
c_plot.figure.set_size_inches(15,5)
c_plot.tick_params(axis='x',rotation=90)
```

```python
# Product Analysis:

# top 10 selling products by quantity

top_10_sell = retail_clean['Description'].value_counts().sort_values(ascending=False).head(10)
top_10_sell
```

```python
top_products = sns.countplot(x='Description',data=retail_clean,order=top_10_sell.index)
top_products.tick_params(axis='x',rotation=85)
```

```python
# top 10 revenue-generating products

# lets make a sub_total columns for easier calculation

retail_clean['Sub_Total'] = retail_clean['Price'] * retail_clean['Quantity']

top_rev = retail_clean.groupby('Description')['Price'].sum().sort_values(ascending=False)
top_rev.head(10)
```
